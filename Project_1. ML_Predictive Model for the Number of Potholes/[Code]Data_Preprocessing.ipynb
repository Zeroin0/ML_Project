# 공휴일 데이터 만들기

## VI. 주중 & 주말 표시하는 column 만들기

```python
from datetime import datetime, date

for j in DF_x['datetime']:
  #print(j)
  i = j.replace('-', ' ')
  d = int(i[:4]), int(i[5:7]), int(i[8:10])
  weekend = date(int(i[:4]), int(i[5:7]), int(i[8:10])).weekday()
  #print(weekend)
  if weekend <= 4:
    DF_x.loc[DF_x['datetime'] == j, 'holiday'] = 'no'
  elif weekend > 4:
    DF_x.loc[DF_x['datetime'] == j, 'holiday'] = 'yes'

DF_x
```

```python
import holidays
years = [2011,2012,2013,2014,2015,2016,2017,2018]
us_holidays = []
for y in years:
  for date in holidays.UnitedStates(years = y).items():
    us_holidays.append(str(date[0]))
  
print(us_holidays)
```

## VIII. 미국 공휴일 holiday column에 표시하기

## IV. 날짜 세부 정보 추출
- 'year', 'month', 'day' 추출

```python
DF.rename(columns={'CREATION DATE':'datetime'}, inplace = True)

DF['datetime'] = DF.datetime.apply(pd.to_datetime)
DF['year'] = DF.datetime.apply(lambda x : x.year)
DF['month'] = DF.datetime.apply(lambda x : x.month)
DF['day'] = DF.datetime.apply(lambda x : x.day)
DF
```

```python
DF.rename(columns={'NUMBER OF POTHOLES FILLED ON BLOCK':'pothole'}, inplace = True)
DF.rename(columns={'icon':'weather'}, inplace = True)

DF = DF.replace('yes', 1)
DF = DF.replace('no', 0)

DF = DF.replace('clear-day', 1)
DF = DF.replace('partly-cloudy-day', 2)
DF = DF.replace('cloudy', 3)
DF = DF.replace('wind', 4)
DF = DF.replace('rain', 5)
DF = DF.replace('snow', 6)

DF = DF[['datetime','year','month','day','holiday','temp','tempmax','tempmin','humidity','precip',
         'snowdepth','solarenergy','weather', 'dew','precipcover', 'windspeed', 'solarradiation','pothole']]

DF
```
